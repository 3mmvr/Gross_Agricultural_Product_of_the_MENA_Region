## **Visualizing Some Of Our Variables To Get Better Understanding Of Our Data**

### Create a chart to see the GAP of the 10 ten performing MENA Countries

```python
gross_agri_prod = full_df[['Country', 'Year', 'Value']]
gross_agri_prod.head()
```

Now the data is much more focused, but not easy to read. Our index is arbitrary and our first column contains tons of duplicates: our dataframe is long when it should be wide. We'll pivot it to make a wide dataframe with a meaningful index and no repeated values.

```python
gross_agri_prod2 = gross_agri_prod.pivot(index='Country', columns='Year', values='Value')
gross_agri_prod2.head()
```

Now we're getting somewhere. Let's get a total of Gross Production over 1961-2007 for each country in the Mena regoin in the dataframe and sort them by that total, then swap the index and columns to prepare for graphing.

```python
gross_agri_prod2.columns = gross_agri_prod2.columns.astype(str)
years = list(map(str, range(1961, 2008)))
gap = gross_agri_prod2
gap['Total'] = gap.sum(axis=1)
gap.sort_values(['Total'], ascending=False, axis=0, inplace=True)
gap = gap[years].transpose()
gap.fillna(0, inplace=True)

gap.head()
```

Now we can seek to list the countries starting with the most agricultural production and ending with the least. This way we can focus on the top 10 best performing countries ini the MENA region and that will also make it easier to graph.

```python
gap_top10 = gap.transpose()
gap_top10 = gap_top10.head(10)
gap_top10_trans = gap_top10.transpose()
gap_top10_trans.head()
```

This Shows that the top 10 Gross Agricultural Producing MENA Countries

```python
gap_top10_trans.fillna(0, inplace=True)
plt.rcParams["figure.autolayout"] = True
ax = gap_top10_trans.plot(kind='area', figsize=(16,8), stacked=False)
ax.set_title('Top Gross Agricultural Producing of MENA Region from 1961 to 2007', fontsize=18, fontweight='bold')
ax.set_ylabel('GAP in Billions (Int $)', fontsize=14)
ax.set_xlabel('')
ax.tick_params(labelsize=13)
```

The area graph show us interesting data in a better visual way than we could've understood from the tables above. The graph shows that Iran has massive growth in its GAP. However, Egypt was performing better than Iran from the 1960s till the mid 1970s, Iran was able to surpass it in terms of GAP with significant value.

Now we've got an area graph of the top 10 gross agricultural producers from 1961 to 2007. A box plot might show us some more about this dataset.

```python
ax0 = gap_top10_trans.plot(kind='box', figsize=(20,8))
ax0.set_title('Top Gross Agricultural Producing MENA Countries from 1961 to 2007', fontsize=20, fontweight='bold')
ax0.set_ylabel('GAP in Billions (Int $)', fontsize=15)
ax0.tick_params(labelsize=10)
```

As we can see Iran and Egypt are at the top of the agricultural Production line. However, other countries between the 2nd and 10th place seem to be highly compatetive and have potential to grow and scale up thier agricultural production.

## Data Modeling using Multivariance Regression & Decision Tree Regression

```python
import sklearn
from sklearn.model_selection import train_test_split # for data split
from sklearn.metrics import explained_variance_score as evs # evaluation metric
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.metrics import mean_squared_error as m2_error
from sklearn.metrics import r2_score as r2 # evaluation metric
from sklearn import preprocessing
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
balance_data = pd.read_csv('cleaned_data.csv',sep= ',', header = None)

new_header = balance_data.iloc[0] #grab the first row for the header
balance_data = balance_data[1:] #take the data less the header row
balance_data.columns = new_header #set the header row as the df header

# Printing the dataswet shape
print ("Dataset Length: ", len(balance_data))
print ("Dataset Shape: ", balance_data.shape)

# Printing the dataset obseravtions
print ("Dataset: ",balance_data.head())
```

```python
X = balance_data.values[:, 3:7].astype(float)
y = balance_data.values[:, 2].astype(int)
```

```python
# creating train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=45)
print(X_train.shape); print(X_test.shape)
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
linear_train_pred = regr.predict(X_train)
print(np.sqrt(m2_error(y_train, linear_train_pred)))
print(r2(y_train, linear_train_pred))
```

```python
linear_test_pred = regr.predict(X_test)
print(np.sqrt(m2_error(y_test, linear_test_pred))) 
print(r2(y_test, linear_test_pred))
```

```python
print("Mean absolute error: %.2f" % np.mean(np.absolute(linear_test_pred - y_test)))
print("Mean sum of squares (MSE): %.2f" % np.mean((linear_test_pred - y_test)** 2))
print("R2-score: %.2f" % r2(linear_test_pred , y_test))
```

```python
#predict the Value of agricultural production of any country given four variables 
#(Arable Land, Area Harvested, Yeild, Production Quantity)

lr_predicted = regr.predict([[7500, 12500000, 32000.0, 31200000]]).astype(int)

# print the predicted price
print("Predicted price: % d\n"% lr_predicted)
```

```python
# import the regressor
from sklearn.tree import DecisionTreeRegressor

# create a regressor object
regressor = DecisionTreeRegressor(random_state=0)

# fit the regressor with X and Y data
regressor.fit(X, y)
```

```python
DRT_train_pred = regressor.predict(X_train)
print('m2_error : ', np.sqrt(m2_error(y_train, DRT_train_pred)))
print('R2_score : ',r2(y_train, DRT_train_pred))
```

```python
DRT_test_pred = regressor.predict(X_test)
print('m2_error : ', np.sqrt(m2_error(y_test, DRT_test_pred))) 
print('R2_score : ', r2(y_test, DRT_test_pred))
```

```python
# model evaluation
print("Mean absolute error: %.2f" % np.mean(np.absolute(DRT_test_pred - y_test)))
print("Mean sum of squares (MSE): %.2f" % np.mean((DRT_test_pred - y_test)** 2))
print("R2-score: %.2f" % r2(DRT_test_pred , y_test))
```

```python
# predicting a new value

# test the output by changing values
dt_predicted = regressor.predict([[7500, 12500000, 32000.0, 31200000]])

# print the predicted price
print("Predicted price: % d\n"% dt_predicted)
```
